```python
import requests
from lxml import html

def extract_people_links(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()  # Raise an error for bad status codes

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Extract href attributes that match the pattern
    people_links = tree.xpath('//a[@class="node-title" and @rel="bookmark"]/@href')

    return people_links

# Example usage:
base_url = 'https://psychology.as.virginia.edu/faculty'  
people_urls = extract_people_links(base_url)

# Print the results
for link in people_urls:
    print(link)
```

```python
def extract_emails(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()  # Raise an error for bad status codes

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Extract href attributes that match the pattern
    emails = tree.xpath('//a[starts-with(@href, "mailto:")]/@href')

    return emails

email_list = []

psych_faculty_url = 'https://psychology.as.virginia.edu/'

for link in people_urls:
    person_url = psych_faculty_url+link
    email_list.append(extract_emails(person_url))

email_list
```

```python
def extract_biography(url):
    response = requests.get(url)
    response.raise_for_status()
    tree = html.fromstring(response.content)

    # Find the parent block that contains <strong>Biography</strong>
    bio_header = tree.xpath('//strong[text()[contains(.,"Biography")]]')
    if not bio_header:
        return None

    parent = bio_header[0].getparent()
    collected = []

    for elem in parent.itersiblings():
        if elem.xpath('.//strong'):
            break  # Stop at next <strong> section
        text = elem.text_content().strip()
        if text:
            collected.append(text)

    return "\n\n".join(collected)

# from lxml import html
# import requests

# def extract_biography(url):
#     response = requests.get(url)
#     response.raise_for_status()
#     tree = html.fromstring(response.content)

#     # First, try <strong>Biography</strong>
#     nodes = tree.xpath('//strong[normalize-space(text())="Biography"]')
#     # If not found, try <h3>Biography</h3>
#     if not nodes:
#         nodes = tree.xpath('//h3[normalize-space(text())="Biography"]')

#     if not nodes:
#         return None

#     header = nodes[0]
#     collected = []

#     # Decide whether to gather from siblings or from following nodes based on the header's tag
#     # If header is <strong>, we look at siblings; if <h3>, often the content is in following <p> tags
#     for elem in header.itersiblings():
#         # Stop collecting if you hit another section heading (<strong> or <h3>)
#         if elem.tag in ['strong', 'h3']:
#             break
#         text = elem.text_content().strip()
#         if text:
#             collected.append(text)

#     return "\n\n".join(collected)


biography_list = []

for link in people_urls:
    person_url = psych_faculty_url+link
    biography_list.append(extract_biography(person_url))

biography_list
```
