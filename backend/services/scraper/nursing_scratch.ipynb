```python
import requests
from lxml import html

def extract_people_links(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()  # Raise an error for bad status codes

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Extract href attributes from anchor tags with the specific class
    people_links = tree.xpath('//a[@class="custom-contact-card-anchor"]/@href')

    return people_links


base_url = 'https://nursing.virginia.edu/research/faculty-research/'  
people_urls = extract_people_links(base_url)

# Print the results
for link in people_urls:
    print(link)
```

```python
email_list = []

def extract_emails(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Extract href values that start with "mailto:" and remove the prefix
    raw_emails = tree.xpath('//a[starts-with(@href, "mailto:")]/@href')
    clean_emails = [email.replace("mailto:", "") for email in raw_emails]

    return clean_emails

for link in people_urls:
    email_list.append(extract_emails(link))
```

```python
import requests
from lxml import html

def extract_biography_text(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Locate the biography section container
    bio_container = tree.xpath('//div[contains(@class, "field--body--biography")]')[0]

    # Extract text only from elements before the first <ul>
    bio_text_parts = []
    for element in bio_container:
        if element.tag == 'ul':
            break
        # Get text content from paragraphs or other tags
        text = element.text_content().strip()
        if text:
            bio_text_parts.append(text)

    return '\n\n'.join(bio_text_parts)

extract_biography_text(people_urls[16])
```
