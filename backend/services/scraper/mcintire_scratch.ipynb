```python
import requests
from lxml import html

def extract_people_links(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()  # Raise an error for bad status codes

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Extract the href attributes from matching <a> elements
    people_links = tree.xpath("//a[starts-with(@href, '/faculty/')]/@href")

    return people_links

# Example usage:
base_url = 'https://www.commerce.virginia.edu/faculty'  # Replace with the actual URL
people_urls = extract_people_links(base_url)

# Print the results
for link in people_urls:
    print(link)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def get_all_faculty_links():
    options = Options()
    options.headless = True  # Run in headless mode (no GUI)
    driver = webdriver.Chrome(options=options)

    try:
        # driver.get("https://www.commerce.virginia.edu/faculty")
        driver.get("https://www.commerce.virginia.edu/faculty?name=&role=faculty")

        # Repeatedly click "Load More" until it's gone or disabled
        while True:
            try:
                load_more_button = WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Load more')]"))
                )
                load_more_button.click()
                time.sleep(1)  # Allow time for content to load
            except Exception:
                break  # No more button or click fails, exit loop

        # Once all faculty are loaded, collect the links
        faculty_elements = driver.find_elements(By.XPATH, "//a[starts-with(@href, '/faculty/')]")
        links = [el.get_attribute("href") for el in faculty_elements]
        return sorted(set(links))  # Remove duplicates and sort

    finally:
        driver.quit()


link_list = []
# Example usage
all_faculty_links = get_all_faculty_links()
for link in all_faculty_links:
    print(link)
    link_list.append(link)
```
