```python
import requests
from lxml import html

def extract_people_links(url):
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()  # Raise an error for bad status codes

    # Parse the HTML
    tree = html.fromstring(response.content)

    # Extract hrefs from <a> elements with rel="bookmark"
    people_links = tree.xpath('//a[@rel="bookmark"]/@href')

    # Optionally prepend the base URL if needed
    full_links = ['https://www.law.virginia.edu' + href for href in people_links if href.startswith('/')]

    return full_links

# Example usage:
base_url = 'https://www.law.virginia.edu/faculty/directory/faculty?faculty_type=521'
people_links = extract_people_links(base_url)

for link in people_links:
    print(link)
```
